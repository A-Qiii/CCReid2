```python
# [Cell 1] 开启自动重载
%load_ext autoreload
%autoreload 2

import torch
from configs import cfg
from datasets import make_dataloader

# 【核心修复】强行读取你硬盘上的 yaml 配置文件
# 注意：请将下面的路径替换为你真实的配置文件路径（比如 'configs/vit_base.yml'）
config_path = 'configs/ltcc/vit_ccreid.yml'
cfg.merge_from_file(config_path)

# 打印出来验明正身，确认路径没有坍缩
print(f"当前生效的数据根目录: {cfg.DATASETS.ROOT_DIR}")
print(f"当前生效的 JSON 路径: {cfg.DATASETS.LLAVA_JSON_PATH}")
```

```python
# [Cell 2] 拦截第一批次的真实数据
train_loader, _, val_loader, num_query, num_classes, camera_num, view_num = make_dataloader(cfg)

batch = next(iter(train_loader))

# 严格按照 bases.py 返回的 7 元组契约进行解包
img, pid, camid, view_id, track_id, id_text, cloth_text = batch

print(f"\n--- 物理矩阵形态 ---")
print(f"图像张量 (img) 形状: {img.shape}")
print(f"身份标签 (pid) 形状: {pid.shape}")

print(f"\n--- 动态组装的文本形态 ---")
print(f"第一张图的 ID Prompt: '{id_text[0]}'")
print(f"第一张图的 衣服 Prompt: '{cloth_text[0]}'")
```

```python
# [合并的完整前向传播测试]

from modeling import make_model

# 1. 重新叫传菜员拿一车全新的菜 (Batch)
batch = next(iter(train_loader))
img, pid, camid, view_id, track_id, id_text, cloth_text = batch

print(f"重新检查批次状态：pid 的类型是 {type(pid)}") # 这里必须是 torch.Tensor 才对！

# 2. 准备显卡和大模型
device = "cuda" if torch.cuda.is_available() else "cpu"
model = make_model(cfg, num_class=num_classes, camera_num=camera_num, view_num=view_num)
model.to(device)
model.train() # 强制开启训练模式

# 3. 将整车数据 (Tensors) 送上显卡
img = img.to(device)
pid = pid.to(device)
camid = camid.to(device)
view_id = view_id.to(device)

# 4. 执行前向传播计算
print("\n>>> 正在执行前向传播计算...")
cls_score, feat_list = model(x=img, label=pid, cam_label=camid, view_label=view_id, 
                             id_text=id_text, cloth_text=cloth_text)

global_feat = feat_list[0]
score_i2t_id = feat_list[1]
score_i2t_cloth = feat_list[2]

print("\n--- 引擎输出的矩阵形态 ---")
print(f"分类得分形状: {cls_score.shape}")
print(f"视觉全局特征形状: {global_feat.shape}")
print(f"图文 ID 相似度矩阵形状: {score_i2t_id.shape}")
print(f"图文衣服相似度矩阵形状: {score_i2t_cloth.shape}")
```

```python
# [Cell 4] 批改试卷：计算分类误差 (Cross Entropy Loss)
import torch.nn.functional as F

# 1. 拿出模型给的答卷 (64张图，每张猜77个人的概率)
prediction = cls_score 

# 2. 拿出真实的正确答案 (64张图真实的主人 ID)
target = pid 

print(f"模型的答卷形状: {prediction.shape}")
print(f"真实的答案形状: {target.shape}")
print(f"看看前3个标准答案的值: {target[:3].tolist()}")

# 3. 核心计算：交叉熵损失 (Cross Entropy)
# 它在底层会去查：这 64 张图，在你标答对应的那个位置上，模型给的预测分到底高不高？不高就产生巨大的 Error。
loss_id = F.cross_entropy(prediction, target)

print(f"\n--- 最终生成的误差惩罚 ---")
print(f"分类误差 (Loss): {loss_id.item():.4f}")
```

```python
# [Cell 5] 触底反弹：误差求导与记忆更新 (Backpropagation)
import torch.optim as optim

# 1. 叫来“修理工” (实例化优化器，对应你 solver 里的机制)
# 我们让它接管模型的所有参数，设定一个微小的学习率
optimizer = optim.Adam(model.parameters(), lr=0.0001)

# 2. 显微镜对焦：随便盯住模型分类器里的一根神经连接 (权重)
# 我们就看第 0 行第 0 列的那个数字
sample_weight = model.classifier.weight[0][0].clone()
print(f"--- 更新前 ---")
print(f"分类器某个权重的当前值: {sample_weight.item():.6f}")
if model.classifier.weight.grad is None:
    print(f"现在的梯度 (修改方向) 是: 空 (还没开始清算)")

# 3. 核心魔法：误差反向传播 (追责开始)
print("\n>>> 正在执行误差反向传播 (loss.backward())...")
optimizer.zero_grad()  # 清空旧账
loss_id.backward()     # 顺着计算图往回骂：计算每个参数该负多少责任！

# 此时，几千万个参数的 .grad 属性全都有了值
grad_value = model.classifier.weight.grad[0][0].item()
print(f"\n--- 算账后 ---")
print(f"它的梯度 (要求修改的方向和力度) 变成了: {grad_value:.6f}")

# 4. 真正长记性：参数物理更新 (往正确的方向走一步)
print("\n>>> 修理工开始调整旋钮 (optimizer.step())...")
optimizer.step()

# 5. 再次用显微镜看这根神经
new_weight = model.classifier.weight[0][0].clone()
print(f"\n--- 更新后 ---")
print(f"分类器这个权重变成了: {new_weight.item():.6f}")
print(f"权重的真实物理形变量: {new_weight.item() - sample_weight.item():.8f}")
```
